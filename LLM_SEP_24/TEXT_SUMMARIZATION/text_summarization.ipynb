### MapReduce chain type

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
docs
```

#### make some more splits

```python

final_docs=RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200).split_documents(docs)
final_docs
```

```python
type(final_docs)
```

```python
len(final_docs)
```

```python
chunk_template=""" 
write a summary of the following speech
Speech:{text}
Summary:
"""
intermediate_prompt=PromptTemplate(input_variables=['text'],template=chunk_template)
```

```python
final_prompt="""
write a short 3 point summary in a table with these points ,and add a motivational title,
Speech:{text}
"""
final_prompt=PromptTemplate(input_variables=['text'],template=final_prompt)
```

```python
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=intermediate_prompt,
    combine_prompt=final_prompt,
    verbose=True
)

output=summary_chain.run(final_docs)
output
```

## refine summerization

```python
chain=load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)
output_summary=chain.run(final_docs)
output_summary
```
